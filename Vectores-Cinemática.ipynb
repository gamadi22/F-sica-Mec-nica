### Ejercicio 1


Considera cuatro vectores en dos dimensiones: $ \mathbf{A} $, $ \mathbf{B} $, $ \mathbf{C} $, y $ \mathbf{D} $.

1. $\mathbf{A} $ tiene una magnitud de $ 10 $ unidades y forma un ángulo de $ 30^\circ $ con el eje $ x $ positivo. Exprésalo en términos de sus componentes cartesianas $ \mathbf{A_x} $ y $ \mathbf{A_y} $.

2. $ \mathbf{B} $ tiene una magnitud de $ 12 $ unidades y forma un ángulo de $ 45^\circ $ con el eje $ y $ positivo. Exprésalo en términos de sus componentes cartesianas $ \mathbf{B_x} $ y $ \mathbf{B_y} $.

3. $ \mathbf{C} $ tiene una magnitud de $ 15 $ unidades y forma un ángulo de $ 60^\circ $ con el eje $ x $ negativo. Exprésalo en términos de sus componentes cartesianas $ \mathbf{C_x} $ y $ \mathbf{C_y} $.

4. $ \mathbf{D} $ tiene una magnitud de $ d = 20 $ unidades y forma un ángulo de $ \theta = -45 $ con el eje $ x $ positivo. Exprésalo en términos de sus componentes cartesianas $ \mathbf{D_x} $ y $ \mathbf{D_y} $ en función de $ d $, $ \theta $, $ x $, e $ y $.

Calcula el vector resultante $ \mathbf{R} $ sumando todos estos vectores y luego determina el ángulo que $ \mathbf{R} $ forma con el eje $ x $ positivo.



```python
import numpy as np

# Definición de los vectores en componentes cartesianas
A_x = 10 * np.cos(np.radians(30))
A_y = 10 * np.sin(np.radians(30))

B_x = 12 * np.cos(np.radians(45))
B_y = 12 * np.sin(np.radians(45))

C_x = -15 * np.cos(np.radians(60))
C_y = 15 * np.sin(np.radians(60))

# Supongamos d y theta
d = 20
theta = np.radians(-45)  # Ángulo en radianes

D_x = d * np.cos(theta)
D_y = d * np.sin(theta)

# Calculamos el vector resultante R
R_x = A_x + B_x + C_x + D_x
R_y = A_y + B_y + C_y + D_y

# Redondeamos a 2 decimales de precisión
R_x = round(R_x, 2)
R_y = round(R_y, 2)

# Calculamos la magnitud de R
magnitude_R = np.sqrt(R_x**2 + R_y**2)

# Calculamos el ángulo que R forma con el eje x
angle_with_x = np.degrees(np.arctan2(R_y, R_x))

# Mostramos la suma de vectores y el ángulo R con el eje X
print("La suma de vectores R es: R_x =", R_x, ", R_y =", R_y)
print("La magnitud de R es:", round(magnitude_R, 2))
print("El ángulo que R forma con el eje x es:", round(angle_with_x, 2), "grados")
```

### Ejercicio 2

Lanzamos una pelota desde una altura de 100 metros con una velocidad inicial de 15 m/s. El ángulo de lanzamiento es de 90 grados en relación a la horizontal. La pelota sube hasta alcanzar la altura máxima y luego cae al suelo. Se nos pide:

    a) Calcular la velocidad de la pelota al llegar al suelo.
    b) Determinar la velocidad de la pelota cuando alcanza la altura máxima.
    c) Calcular la aceleración cuando la pelota alcanza la altura máxima.
    d) Calcular la posición de la pelota cuando han transcurrido 1 segundo, 5 segundos y 10 segundos.

Vamos a abordar cada una de las preguntas paso a paso para este nuevo escenario en el que el ángulo de lanzamiento es de 90 grados en relación a la horizontal.

**a) Calcular la velocidad de la pelota al llegar al suelo**

Dado que el ángulo de lanzamiento es de 90 grados (lanzamiento vertical hacia arriba), la velocidad final al llegar al suelo será igual a la velocidad inicial, pero con signo negativo debido al sentido hacia abajo. Por lo tanto, la velocidad de la pelota al llegar al suelo será $ -15 \, \text{m/s} $ (15 m/s hacia abajo).

**b) Determinar la velocidad de la pelota cuando alcanza la altura máxima**

Como la velocidad inicial es de 15 m/s hacia arriba, en el punto más alto de la trayectoria (altura máxima), la velocidad será cero. La componente vertical de la velocidad inicial se reduce debido a la aceleración debida a la gravedad hasta que se detiene por completo antes de cambiar de dirección y volver hacia abajo.

**c) Calcular la aceleración cuando la pelota alcanza la altura máxima**

La aceleración debido a la gravedad actúa constantemente hacia abajo, pero en el punto más alto del lanzamiento su velocidad vertical llega a cero antes de cambiar de dirección y volver hacia abajo. Por lo tanto, en el punto más alto, la aceleración debido a la gravedad es $ g \approx 9.81 \, \text{m/s}^2 $, y esta es la aceleración en ese punto.

**d) Calcular la posición de la pelota cuando han transcurrido 1 segundo, 5 segundos y 10 segundos**

Para calcular la posición de la pelota en estos momentos, utilizaremos la misma ecuación de la cinemática que mencioné anteriormente:

$ y(t) = y_0 + v_{0y}t - \frac{1}{2}gt^2 $

Donde:
- $ y(t) $ es la posición en un momento dado $ t $.
- $ y_0 $ es la altura inicial.
- $ v_{0y} $ es la componente inicial de la velocidad en la dirección vertical.
- $ g $ es la aceleración debido a la gravedad.

Ahora, vamos a calcular la posición en los momentos dados:

1. **Cuando han transcurrido 1 segundo ($ t = 1 $ segundo):**
   - $ y(1) = 0 \, \text{m} + (15 \, \text{m/s})(1 \, \text{s}) - \frac{1}{2}(9.81 \, \text{m/s}^2)(1 \, \text{s})^2 $

2. **Cuando han transcurrido 5 segundos ($ t = 5 $ segundos):**
   - $ y(5) = 0 \, \text{m} + (15 \, \text{m/s})(5 \, \text{s}) - \frac{1}{2}(9.81 \, \text{m/s}^2)(5 \, \text{s})^2 $

3. **Cuando han transcurrido 10 segundos ($ t = 10 $ segundos):**
   - $ y(10) = 0 \, \text{m} + (15 \, \text{m/s})(10 \, \text{s}) - \frac{1}{2}(9.81 \, \text{m/s}^2)(10 \, \text{s})^2 $



```python
def calcular_posicion(tiempo, velocidad_inicial, gravedad):
    # Calcula la posición en un momento dado
    posicion = velocidad_inicial * tiempo - 0.5 * gravedad * tiempo ** 2
    return posicion

# Datos del problema
altura_inicial = 0  # Altura desde la que se lanza la pelota (m)
velocidad_inicial = 15  # Velocidad inicial (m/s)
gravedad = 9.81  # Aceleración debido a la gravedad (m/s^2)

# Puntos de tiempo para los que queremos calcular la posición
tiempos = [1, 5, 10]  # Tiempos en segundos

# a) Calcular la velocidad de la pelota al llegar al suelo
velocidad_al_suelo = -velocidad_inicial  # La velocidad final al llegar al suelo

# b) Determinar la velocidad de la pelota cuando alcanza la altura máxima
velocidad_maxima = 0  # La velocidad es cero en el punto más alto

# c) Calcular la aceleración cuando la pelota alcanza la altura máxima
aceleracion_maxima = gravedad  # La aceleración debido a la gravedad

# d) Calcular la posición de la pelota en los tiempos especificados
for tiempo in tiempos:
    posicion = calcular_posicion(tiempo, velocidad_inicial, gravedad)
    print(f'Posición en t = {tiempo} segundos: {posicion} metros')

print(f'Velocidad al llegar al suelo: {velocidad_al_suelo} m/s')
print(f'Velocidad en la altura máxima: {velocidad_maxima} m/s')
print(f'Aceleración en la altura máxima: {aceleracion_maxima} m/s^2')
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Funciones para calcular posición y velocidad
def calcular_posicion(tiempo, velocidad_inicial, gravedad):
    return velocidad_inicial * tiempo - 0.5 * gravedad * tiempo ** 2

def calcular_velocidad(tiempo, velocidad_inicial, gravedad):
    return velocidad_inicial - gravedad * tiempo

# Datos del problema
altura_inicial = 0  # Altura desde la que se lanza la pelota (m)
velocidad_inicial = 15  # Velocidad inicial (m/s)
gravedad = 9.81  # Aceleración debido a la gravedad (m/s^2)

# Generar valores de tiempo
tiempos = np.linspace(0, 10, 100)  # Tiempo de 0 a 10 segundos, 100 puntos

# Calcular posición y velocidad en función del tiempo
posiciones = calcular_posicion(tiempos, velocidad_inicial, gravedad)
velocidades = calcular_velocidad(tiempos, velocidad_inicial, gravedad)

# Crear las gráficas
plt.figure(figsize=(12, 5))

# Gráfica de posición en función del tiempo
plt.subplot(1, 2, 1)
plt.plot(tiempos, posiciones)
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.title('Posición de la pelota en función del tiempo')

# Gráfica de velocidad en función del tiempo
plt.subplot(1, 2, 2)
plt.plot(tiempos, velocidades)
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.title('Velocidad de la pelota en función del tiempo')

plt.tight_layout()
plt.show()
```

Supongamos que estamos en un planeta cuya aceleración debida a la gravedad $ g $ varía con la altura $ h $ desde la superficie. La variación se describe mediante la ecuación:

$ g(h) = g_0 \left( \frac{R}{R + h} \right)^2 $

Donde:
- $ g_0 $ es la aceleración debida a la gravedad en la superficie del planeta.
- $ R $ es el radio del planeta.

Queremos calcular la velocidad de escape desde la superficie del planeta.

**Preguntas:**

a) ¿Cuál es la aceleración debida a la gravedad en la superficie del planeta ($ h = 0 $)?

b) Calcule la aceleración debida a la gravedad a una altura de $ h = 2R $.

c) Determine la velocidad de escape desde la superficie del planeta.

**Datos:**
- Aceleración debida a la gravedad en la superficie del planeta ($ h = 0 $): $ g_0 = 9.81 \, \text{m/s}^2 $
- Radio del planeta: $ R = 6371 \, \text{km} $ (radio aproximado de la Tierra)


```python
# Datos
g0 = 9.81  # Aceleración debida a la gravedad en la superficie del planeta (m/s^2)

# Cálculo de g0 (aceleración en la superficie)
g_surface = g0
print(f"Aceleración en la superficie del planeta (h = 0): {g_surface} m/s^2")
```

Primero, simplificamos la ecuación:

$ 2v_{A0}t + a_A t^2 = 2d_0 + 2v_{B0}t + a_{Bf}t^2 - 4a_{Bf}t + 4a_{Bf} $

Organizamos términos:

$ (a_A - a_{Bf})t^2 + (2v_{A0} - 2v_{B0} - 4a_{Bf})t + (2d_0 - 4a_{Bf} + 4a_{Bf}) = 0 $

Simplificamos:

$ (a_A - a_{Bf})t^2 + (2v_{A0} - 2v_{B0} - 4a_{Bf})t + 2d_0 = 0 $

Ahora, podemos aplicar la fórmula cuadrática para resolver para \(t\):

$ t = \frac{-(2v_{A0} - 2v_{B0} - 4a_{Bf}) \pm \sqrt{(2v_{A0} - 2v_{B0} - 4a_{Bf})^2 - 4(a_A - a_{Bf}) \cdot 2d_0}}{2(a_A - a_{Bf})} $

Esta fórmula proporcionará dos soluciones para \(t\), y debemos elegir la solución que tenga sentido en el contexto del problema.

```python
# Datos
v_A0 = 10  # Velocidad inicial del vehículo A en m/s
v_B0 = 15  # Velocidad inicial del vehículo B en m/s
a_A = 2    # Aceleración del vehículo A en m/s^2
a_B = 1.5  # Aceleración del vehículo B en m/s^2
d0 = 100   # Distancia inicial entre los vehículos en metros

# a) Calcular el tiempo de encuentro (en segundos)
t_encuentro = (-2 * d0) / (v_A0 + v_B0)  # Usando la ecuación de encuentro
print(f"Tiempo de encuentro de los vehículos: {t_encuentro:.2f} segundos")
```

```python
# b) Calcular la distancia recorrida por cada vehículo en el momento del encuentro
distancia_recorrida_A = v_A0 * t_encuentro + 0.5 * a_A * t_encuentro**2
distancia_recorrida_B = v_B0 * t_encuentro + 0.5 * a_B * t_encuentro**2

print("Distancia recorrida por el vehículo A en el encuentro:", distancia_recorrida_A, "metros")
print("Distancia recorrida por el vehículo B en el encuentro:", distancia_recorrida_B, "metros")
```

```python
# Datos para el vehículo que frena
a_f = -10  # Deceleración debido al frenado en m/s^2
t_frenado = -v_B0 / a_f  # Tiempo en el que la velocidad se reduce a 0 (cambio de signo)

# Calcular la distancia recorrida por el vehículo que frena hasta frenar
distancia_recorrida_frenado = v_B0 * t_frenado + 0.5 * a_f * t_frenado**2

# Calcular el tiempo de encuentro considerando el frenado
t_encuentro_con_frenado = t_encuentro - t_frenado  # Tiempo de encuentro ajustado

print("Tiempo en que el vehículo B comienza a frenar:", t_frenado, "segundos")
print("Distancia recorrida por el vehículo B durante el frenado:", distancia_recorrida_frenado, "metros")
print("Tiempo de encuentro ajustado considerando el frenado:", t_encuentro_con_frenado, "segundos")
```

### Ejercicio 6

En este ejercicio, consideraremos el movimiento de una barca en el mar, donde la aceleración de la barca varía con el tiempo siguiendo una función senoidal. Vamos a determinar la posición y velocidad de la barca en función del tiempo, y luego calcular la velocidad y aceleración en función del espacio recorrido. Finalmente, representaremos gráficamente la velocidad, aceleración y posición en función del tiempo, así como la aceleración y velocidad en función del espacio.

**Parte 1: Movimiento en función del tiempo**

a) Dada la aceleración $a(t)$ de la barca en función del tiempo $t$, determinar la velocidad $v(t)$ y la posición $x(t)$ en función del tiempo $t$.

La aceleración de la barca sigue una función senoidal, por ejemplo:

$ a(t) = A \sin(\omega t) $

Donde:
- $ A $ es la amplitud de la aceleración.
- $ \omega $ es la frecuencia angular.

**Parte 2: Movimiento en función del espacio recorrido**

b) Si se nos da el espacio recorrido $s$, calcular la velocidad $v(s)$ y la aceleración $a(s)$ en función del espacio recorrido $s$.

**Parte 3: Representación gráfica**

c) Representar gráficamente la velocidad, aceleración y posición en función del tiempo $t$.

d) Representar gráficamente la aceleración y velocidad en función del espacio recorrido $s$.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
A = 5.0  # Amplitud de la aceleración
omega = 2.0  # Frecuencia angular

# Tiempo
t = np.linspace(0, 10, 400)  # Generar 400 puntos entre 0 y 10 segundos

# Aceleración
a = A * np.sin(omega * t)

# Velocidad
v = (A / omega) * (1 - np.cos(omega * t))

# Posición
x = (A / (omega ** 2)) * (omega * t - np.sin(omega * t))

# Gráficos
plt.figure(figsize=(12, 10))

# Aceleración en función del tiempo
plt.subplot(3, 1, 1)
plt.plot(t, a, label='Aceleración (a)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Aceleración (m/s^2)')
plt.legend()
plt.grid()

# Velocidad en función del tiempo
plt.subplot(3, 1, 2)
plt.plot(t, v, label='Velocidad (v)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Velocidad (m/s)')
plt.legend()
plt.grid()

# Posición en función del tiempo
plt.subplot(3, 1, 3)
plt.plot(t, x, label='Posición (x)')
plt.xlabel('Tiempo (s)')
plt.ylabel('Posición (m)')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
# Cálculo numérico de v(s) en función de s
def calcular_velocidad_s(s_values):
    v_s = np.zeros_like(s_values)
    for i, s in enumerate(s_values):
        # Encontrar el tiempo correspondiente a la posición s
        tiempo = np.interp(s, x, t)
        # Calcular la velocidad correspondiente
        v_s[i] = (A / omega) * (1 - np.cos(omega * tiempo))
    return v_s

# Espacio recorrido (s) en función del tiempo
s = x

# Calcular la velocidad en función del espacio
v_s = calcular_velocidad_s(s)

# Gráfico de la velocidad en función del espacio
plt.figure(figsize=(8, 6))
plt.plot(s, v_s, label='Velocidad (v)')
plt.xlabel('Espacio recorrido (s) (m)')
plt.ylabel('Velocidad (m/s)')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calcular la aceleración en función del espacio (derivada de v(s))
a_s = np.gradient(v_s, s)

# Gráfico de la aceleración en función del espacio
plt.figure(figsize=(8, 6))
plt.plot(s, a_s, label='Aceleración (a)')
plt.xlabel('Espacio recorrido (s) (m)')
plt.ylabel('Aceleración (m/s^2)')
plt.legend()
plt.grid()
plt.show()
```
