### Movimiento Parabólico y caída libre

Demostraremos que dado un ángulo y velocidades iniciales en un movimiento parabólico, siempre que estemos dentro del radio de acción del movimiento, se puede demostrar que proyectil lanzado con tiro parabólico y la caída de un objeto en el mismo instante, siempre, siempre, siempre chocarán... Para ello, es necesario que se encuentren en línea.

```python
from vpython import *
#GlowScript 3.1 VPython

# valores iniciales (modifícalos bajo tu responsabilidad)

v0 = 20 ## v en m/s
alfa = 60 ## ángulo en grados
alfa_rad = alfa * pi/180
vel_fotogramas = 20  ## la velocidad del video
porcentaje = 0.5 ## relación de posición en relación a x_final

# Constantes
g = 9.81 ## Aceleración de la gravedad

print("Introduce la distancia relativa del movimiento de caída libre (número entre 0 y 1): ")
porcentaje = input()

print("Introduce el ángulo de disparo (número entre 0 y 90): ")
alfa = input()
alfa_rad = float(alfa) * pi/180

# ecuaciones
v0x = v0 * cos(alfa_rad)
v0y = v0 * sin(alfa_rad)

## v0y = 1/2 * g * t_total
t_total = 2 * v0y / g
x_final = v0x * t_total
d = x_final * float(porcentaje)
print(d)

# Empezamos con visual python (vpython)
# Creamos el 'suelo'
suelo = box(
    pos=vector(x_final/2, -1, 0),
    size=vector(x_final, 1, 30), 
    color=color.green
)

# Creamos el 'cañón'
canyon = cylinder(
    pos=vector(0, 0, 0),
    axis=vector(
        2 * cos(alfa_rad), 
        2 * sin(alfa_rad),
        0
    )
)

# Creamos el proyectil y una línea que dejará la estela del proyectil
bola = sphere(pos=vector(0, 0, 0))
bola.trail = curve(color=bola.color)

# Creamos la bola de caída libre
ball = sphere(pos = vector(d,0,0), radius = 1, color = color.green) 


# Creamos la flecha que indica la dirección del movimiento (vector velocidad)
flecha = arrow(
    pos=vector(0, 0, 0),
    axis=vector(v0x, v0y, 0), 
    color=color.yellow
)
# texto (ponemos etiquetas para informar de la posición del proyectil)
labelx = label(
    pos=bola.pos,
    text='posicion x = 0 m',
    xoffset=1,
    yoffset=80,
    space=bola.radius,
    font='sans',
    box=False,
    height=10,
)

labely = label(
    pos=bola.pos,
    text='posicion y = 0 m',
    xoffset=1,
    yoffset=40,
    space=bola.radius,
    font='sans',
    box=False,
    height=10
)
# Animamos todo el cotarro!!!
t = 0
# Condiciones iniciales de la bola en caída libre
v0 = 0
t0 = 0
y0 = d*tan(alfa_rad)
y = y0

while t <= t_total:
    bola.pos = vector(
        v0x * t, 
        v0y * t - 0.5 * g * t**2,
        0)
    flecha.pos = vector(
        v0x * t, 
        v0y * t - 0.5 * g * t**2,
        0
    )
    if (y>0):
        y = y0-0.5 *g*t**2+v0*t
    
    ball.pos=vec(d,y,0)
    
    flecha.axis = vector(v0x, v0y - g * t, 0)
    bola.trail.append(pos=bola.pos)
    labelx.pos = bola.pos
    labelx.text = 'posicion x = ' + str(v0x * t)
    labely.pos = bola.pos
    labely.text = 'posicion y = ' + str(v0y * t - 0.5 * g * t**2)
#    t = t+0.01
    t = t + t_total / 100.
    rate(vel_fotogramas)
```

### Péndulo

```python
from vpython import *

#GlowScript 2.9 VPython

display(width=600,height=600,center=vector(0,12,0),background=color.white)

g=9.8 # acelaración de la gravedad

bob=sphere(pos=vector(5,2,0),radius=0.5,color=color.blue)

pivot=vector(0,20,0)
roof=box(pos=pivot,size=vector(10,0.5,10),color=color.green)
rod=cylinder(pos=pivot,axis=bob.pos-pivot,radius=0.1,color=color.red)

t=0 # tiempo inicial 
dt=0.01 # diferencial de tiempo 

l=mag(bob.pos-pivot) # Longitud del péndulo
cs=(pivot.y-bob.pos.y)/l # Calculando cos(theta) 
theta=acos(cs) # Ángulo con la vertical
vel=0.0 # Velocidad angular

while (t<100):
    rate(100) # máximo 100 cálculos por segundo
    acc=-g/l*sin(theta) # actualización de la aceleración angular
    theta=theta+vel*dt # actualización de la posición angular
    vel=vel+acc*dt # actualización de la velocidad angular
    bob.pos=vector(l*sin(theta),pivot.y-l*cos(theta),0) # cálculo de la posición
    rod.axis=bob.pos-rod.pos # actualización del otro extremo de la varilla del péndulo
    t=t+dt # tiempo de actualización
```

### Tiro parabólico con rozamiento

```python
from vpython import *
#GlowScript 3.1 VPython
## Proyectil con rozamiento.

## 0.5*C*rho*Area*mag2(ball.v)*ball.v.hat 
## 

scene.center = vector(50,20,-10)
scene.fov = pi/2.5

box(pos=vector(50,0,0), size=vector(100,0.1,5), color=color.yellow)

ball = sphere(pos=vector(0,0,0), radius=0.2, color=color.green, make_trail=True)

speed = 65
angle = 45/180*pi

vx = speed * cos(angle)
vy = speed * sin(angle)
ball.v = vector(vx,vy,0)

m=2.0; rad=0.2; g=9.8
dt = 0.001
rho = 1.2; C=0.5; Area = pi*rad**2

while ball.pos.y > -0.01:
    rate(400)
    F=vector(0,-m*g,0) - 0.5*C*rho*Area*mag2(ball.v)*ball.v.hat
    a=F/m
    ball.v=ball.v+a*dt
    ball.pos=ball.pos+ball.v*dt
print("El alcance es de: ", ball.pos.x, "metros.")
```

### Sistema Solar

```python
from vpython import *
#GlowScript 3.1 VPython

scene.title = "Modelización del movimiento de planetas por la fuerza gravitatoria"
scene.height = 600
scene.width = 800


star = sphere(pos=vector(0,0,0), radius=0.2, color=color.yellow,
               mass = 1000, momentum=vector(0,0,0), make_trail=True)

planet1 = sphere(pos=vector(1,0,0), radius=0.05, color=color.green,
               mass = 1, momentum=vector(0,30,0), make_trail=True)

planet2 = sphere(pos=vector(0,3,0), radius=0.075, color=vector(0.0,0.82,0.33),#RGB color
                  mass = 2, momentum=vector(-35,0,0), make_trail=True)
                  
planet3  = sphere(pos=vector(0,-4,0), radius=0.1, color=vector(0.58,0.153,0.68),
                  mass = 10, momentum=vector(160,0,0), make_trail=True)

#real-world value is : G = 6.67e-11

def gravitationalForce(p1,p2):
    G=1
    rVector = p1.pos - p2.pos
    rMagnitude = mag(rVector)
    rHat = rVector / rMagnitude
    F = -rHat * G * p1.mass * p2.mass / rMagnitude**2
    return F

t = 0
dt = 0.0001 #El tamaño del paso. Debe ser un número pequeño

while (True):
    rate(500)
    star.force = gravitationalForce(star,planet1)+gravitationalForce(star,planet2)+gravitationalForce(star,planet3)
    planet1.force = gravitationalForce(planet1,star)+gravitationalForce(planet1,planet2)+gravitationalForce(planet1,planet3)
    planet2.force = gravitationalForce(planet2,star)+gravitationalForce(planet2,planet1)+gravitationalForce(planet2,planet3)
    planet3.force = gravitationalForce(planet3,star)+gravitationalForce(planet3,planet1)+gravitationalForce(planet3,planet2)
    
    star.momentum = star.momentum + star.force*dt
    planet1.momentum = planet1.momentum + planet1.force*dt
    planet2.momentum = planet2.momentum + planet2.force*dt
    planet3.momentum = planet3.momentum + planet3.force*dt
    
    star.pos = star.pos + star.momentum/star.mass*dt
    
    planet1.pos = planet1.pos + planet1.momentum/planet1.mass*dt
    planet2.pos = planet2.pos + planet2.momentum/planet2.mass*dt
    planet3.pos = planet3.pos + planet3.momentum/planet3.mass*dt
    t += dt
    
```
