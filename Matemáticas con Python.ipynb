## Ecuaciones Diferenciales

### Ecuaciones diferenciales lineales

Uno de los tipos ms importantes de Ecuaciones diferenciales son las Ecuaciones diferenciales lineales. Este tipo de ecuaciones son muy comunes en varias ciencias y tienen la ventaja de que pueden llegar a ser resueltas en forma analtica ya que su ecuacin diferencial de primer orden adopta la forma:

$\frac{dy}{dx} + P(x)y = Q(x)$

Para resolver este tipo de ecuaciones de la forma $y' + P(x)y = Q(x)$, debemos multiplicar los dos lados de la ecuacin por el factor de integracin $e^{\int P(x) dx}$ y luego integrar ambos lados. As, si por ejemplo quisiramos resolver la siguiente Ecuacin diferencial: 
$\frac{dy}{dx} + 3x^2y = 6x^2$

Primero calculamos el factor de integracin, $I(x) = e^{\int 3x^2 \ dx} = e^{\int 3x^2 \ dx}$

Luego multiplicamos ambos lados de la ecuacin por el recin calculado factor de integracin.

$e^{x^3}\frac{dy}{dx} + 3x^2e^{x^3}y = 6x^2e^{x^3}$, que simplificando obtenemos: 

$\frac{d}{dx}(e^{x^3}y) = 6x^2e^{x^3}$

$e^{x^3}y = \int 6x^2e^{x^3} \ dx = 2e^{x^3} + C$

$y = Ce^{-x^3} +2$

#### Condicin Inicial
Cuando utilizamos Ecuaciones diferenciales, generalmente no estamos interesados en encontrar una familia de soluciones (la solucin general), como la que hayamos en el caso precedente, sino que vamos a estar ms interesados en una solucin que satisface un requerimiento particular.En muchos problemas fsicos debemos de encontrar una solucin particular que satisface una condicin de la forma $y(t_o) = y_0$. Esto se conoce como la condicin inicial, y el problema de encontrar una solucin de la Ecuacin diferencial que satisface la condicin inicial se conoce como el problema de valor inicial.

### Series de potencias

Cuando comenzamos a lidiar con las Ecuaciones diferenciales, veremos que existen un gran nmero de ellas que no pueden ser resueltas en forma analtica utilizando los principios del Clculo integral y el Clculo diferencial; pero sin embargo, tal vez podamos encontrar soluciones aproximadas para este tipo de ecuaciones en trminos de Series de potencias.

$f(x) = C_0 + C_1X + C_2X^2 + C_3X^3 + \dots + C_nX^n + \dots$

El dominio de esta funcin va a estar dado por el conjunto de todos los X para los que la serie converge

#### Series de Taylor

Una de las razones de que las Series de Taylor sean importantes es que nos permiten integrar funciones que de otra forma no podamos manejar. De hecho, a menudo Newton integraba funciones por medio de, en primer lugar, expresarlas como Series de potencias y luego integrando la serie trmino a trmino. 

Por ejemplo, la funcin $f(x) = e^{-x^2}$, no puede ser integrada por los mtodos normales del Clculo integral, por lo que debemos recurrir a las Series de Taylor para aproximar la solucin de su integral. Podramos construir la Serie de Taylor de esta funcin, del siguiente modo:

$e^{-x^2} = \sum_{n=0}^{\infty}\frac{(-x^2)^n}{n!} = \sum_{n=0}^{\infty}(-1)^n\frac{x^{2n}}{n!} =
1 - \frac{x^2}{1!} + \frac{x^4}{2!} -\frac{x^6}{3!} + \dots$

#### Soluciones analticas con Python

SymPy nos proporciona un solucionador genrico de Ecuaciones diferenciales ordinarias, sympy.dsolve, el cual es capaz de encontrar soluciones analticas a muchas EDOs elementales. Mientras sympy.dsolve se puede utilizar para resolver muchas EDOs sencillas simblicamente, como veremos a continuacin, debemos tener en cuenta que la mayora de las EDOs no se pueden resolver analticamente. Por ejemplo, retomando el ejemplo que resolvimos analticamente ms arriba, veamos si llegamos al mismo resultado utilizando SymPy para solucionar la siguiente Ecuacin diferencial ordinaria:

$\frac{dy}{dx} = -3x^2y + 6x^2$

```python
# Resolviendo ecuacin diferencial
# defino las incgnitas
import sympy

sympy.init_printing(use_latex = 'mathjax')

x = sympy.Symbol('x')
y = sympy.Function('y')

# expreso la ecuacin
f = 6*x**2 - 3*x**2*(y(x))
sympy.Eq(y(x).diff(x), f)
```

```python
sympy.dsolve(y(x).diff(x) - f)
```

### Transformada de Laplace
Un mtodo alternativo que podemos utilizar para resolver en forma analtica Ecuaciones diferenciales ordinarias complejas, es utilizar la Transformada de Laplace, que es un tipo particular de transformada integral. La idea es que podemos utilizar esta tcnica para transformar nuestra Ecuacin diferencial en algo ms simple, resolver esta ecuacin ms simple y, a continuacin, invertir la transformacin para recuperar la solucin a la Ecuacin diferencial original.

Si tenemos la siguiente ecuacin: $y'' + 3y' + 2y = 0$

Con las siguientes condiciones iniciales, $y(0) = 2$; $y'(0) = -3$

```python
# Ejemplo de transformada de Laplace
# Defino las incgnitas
t = sympy.symbols("t", positive=True)
y = sympy.Function("y")

# Defino la ecuacin
edo = y(t).diff(t, t) + 3*y(t).diff(t) + 2*y(t)
sympy.Eq(edo)
```

```python
# smbolos adicionales.
s, Y = sympy.symbols("s, Y", real=True)
```

```python
def laplace_transform_derivatives(e):
    """
    Evala las transformadas de Laplace de derivadas de funciones sin evaluar.
    """
    if isinstance(e, sympy.LaplaceTransform):
        if isinstance(e.args[0], sympy.Derivative):
            d, t, s = e.args 
            n = len(d.args) - 1
            return ((s**n) * sympy.LaplaceTransform(d.args[0], t, s) -
                    sum([s**(n-i) * sympy.diff(d.args[0], t, i-1).subs(t, 0)
                         for i in range(1, n+1)]))
        
    if isinstance(e, (sympy.Add, sympy.Mul)):
        t = type(e) 
        return t(*[laplace_transform_derivatives(arg) for arg in e.args])
    
    return e
```

```python
# Clculo la transformada de Laplace 
L_edo = sympy.laplace_transform(edo, t, s)
sympy.Eq(L_edo)
```

```python
# Aplicamos la nueva funcin para evaluar las transformadas de Laplace
# de derivadas
L_edo_2 = laplace_transform_derivatives(L_edo)
sympy.Eq(L_edo_2)
```

### Movimiento con aceleracin constante

En un movimiento acelerado con aceleracin constante, las ecuaciones de velocidad y espacio pueden deducirse utilizando la cinemtica bsica. Supongamos que un objeto se mueve con aceleracin constante $a$ y que en $t = 0$ se encuentra en $x = x_0$ con velocidad inicial $v_0$.

#### 1. Ecucin de Velocidad:

La ecuacin de velocidad en funcin del tiempo se deduce de la definicin de aceleracin como la tasa de cambio de la velocidad:

$ a = \frac{dv}{dt} $

Integrando esta ecuacin con respecto al tiempo, obtendremos la ecuacin de velocidad:

$ \int a \, dt = \int \frac{dv}{dt} \, dt $

$ \int a \, dt = v - v_0 $

Donde $v_0$ es la velocidad inicial, y $v$ es la velocidad en cualquier momento $t$.

Entonces, la ecuacin de velocidad es:

$ v = v_0 + at $

#### 2. Ecuacin de Espacio:

La ecuacin de espacio en funcin del tiempo se deriva de manera similar. Comenzamos con la ecuacin de velocidad:

$ v = v_0 + at $

Luego, utilizamos la definicin de velocidad promedio como la tasa de cambio de la posicin:

$ v_{\text{promedio}} = \frac{\Delta x}{\Delta t} $

Donde $\Delta x$ es el cambio en la posicin y $\Delta t$ es el cambio en el tiempo.

Podemos reorganizar esta ecuacin para despejar $\Delta x$:

$ \Delta x = v_{\text{promedio}} \cdot \Delta t $

Ahora, consideramos que durante un intervalo de tiempo infinitesimal $dt$, la velocidad vara desde $v$ hasta $v + dv$, y el espacio vara desde $x$ hasta $x + dx$. Entonces, podemos expresar el espacio en funcin del tiempo y la velocidad:

$ dx = v \, dt $

Finalmente, sustituimos la ecuacin de velocidad $v = v_0 + at$ en esta expresin:

$ dx = (v_0 + at) \, dt $

Integrando esta ecuacin con respecto al tiempo, obtenemos la ecuacin de espacio:

$ \int dx = \int (v_0 + at) \, dt $

$ x - x_0 = v_0t + \frac{1}{2}at^2 $

Donde $x_0$ es la posicin inicial y $x$ es la posicin en cualquier momento $t$.

Entonces, la ecuacin de espacio es:

$ x = x_0 + v_0t + \frac{1}{2}at^2 $

Estas son las ecuaciones de velocidad y espacio para un movimiento acelerado con aceleracin constante. Puedes utilizar estas ecuaciones para describir el movimiento de un objeto bajo estas condiciones.

### Movimiento con aceleracin variable

Para el caso de un movimiento con aceleracin dependiente del tiempo, podemos deducir las ecuaciones de velocidad y espacio de manera similar, pero teniendo en cuenta que la aceleracin vara con el tiempo $t$. Supongamos que la aceleracin en funcin del tiempo es $a(t)$.

#### 1. Ecuacin de Velocidad:

Comenzamos con la definicin de aceleracin:

$ a(t) = \frac{dv}{dt} $

Integrando esta ecuacin con respecto al tiempo, obtendremos la ecuacin de velocidad:

$ \int a(t) \, dt = \int \frac{dv}{dt} \, dt $

$ \int a(t) \, dt = v - v_0 $

Donde $v_0$ es la velocidad inicial, y $v$ es la velocidad en cualquier momento $t$.

Entonces, la ecuacin de velocidad es:

$ v(t) = v_0 + \int a(t) \, dt $

La integral $\int a(t) \, dt$ representa la acumulacin de cambios de velocidad en funcin del tiempo y se obtiene evaluando la integral definida en el intervalo de tiempo apropiado.

#### 2. Ecuacin de Espacio:

La ecuacin de espacio en funcin del tiempo se deriva de manera similar. Comenzamos con la ecuacin de velocidad:

$ v(t) = v_0 + \int a(t) \, dt $

Luego, utilizamos la definicin de velocidad promedio como la tasa de cambio de la posicin:

$ v_{\text{promedio}} = \frac{\Delta x}{\Delta t} $

Donde $\Delta x$ es el cambio en la posicin y $\Delta t$ es el cambio en el tiempo.

Podemos reorganizar esta ecuacin para despejar $\Delta x$:

$ \Delta x = v_{\text{promedio}} \cdot \Delta t $

Ahora, consideramos que durante un intervalo de tiempo infinitesimal $dt$, la velocidad vara desde $v(t)$ hasta $v(t) + dv(t)$, y el espacio vara desde $x(t)$ hasta $x(t) + dx(t)$. Entonces, podemos expresar el espacio en funcin del tiempo y la velocidad:

$ dx(t) = v(t) \, dt $

Finalmente, sustituimos la ecuacin de velocidad $v(t) = v_0 + \int a(t) \, dt$ en esta expresin:

$ dx(t) = \left(v_0 + \int a(t) \, dt\right) \, dt $

Integrando esta ecuacin con respecto al tiempo, obtenemos la ecuacin de espacio:

$ \int dx(t) = \int \left(v_0 + \int a(t) \, dt\right) \, dt $

$ x(t) - x_0 = v_0t + \int \left(\int a(t) \, dt\right) \, dt $

Donde $x_0$ es la posicin inicial y $x(t)$ es la posicin en cualquier momento $t$.

Entonces, la ecuacin de espacio es:

$ x(t) = x_0 + v_0t + \int \left(\int a(t) \, dt\right) \, dt $

En este caso, la integral doble $\int \left(\int a(t) \, dt\right) \, dt$ representa la acumulacin de cambios de velocidad en funcin del tiempo y se obtiene evaluando la integral definida en el intervalo de tiempo apropiado.

Estas son las ecuaciones de velocidad y espacio para un movimiento con aceleracin dependiente del tiempo. Puedes utilizar estas ecuaciones para describir el movimiento de un objeto bajo estas condiciones, evaluando las integrales en funcin de la aceleracin en funcin del tiempo $a(t)$.
